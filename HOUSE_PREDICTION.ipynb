```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #used in finding the correlation
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor #using xgbooster here
from sklearn import metrics

#importing data directly from sklearn
house_price_dataset=sklearn.datasets.load_boston()
print(house_price_dataset)
```

```python
#making datsets in pandas frame
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
house_price_dataframe.head(10)
```

```python
#or we can add data manually
#house_price_dataset_1=pd.read_csv('/content/housing.csv',header=None)
#house_price_dataset_1.head()
```

```python
#adding the target value to the data
#since the data is not manually uploaded target value will be removed 
#so add the target value
house_price_dataframe['price'] = house_price_dataset.target
house_price_dataframe.head()
```

```python
#checking the dimensions
house_price_dataframe.shape
```

```python
#finding the missing values
house_price_dataframe.isnull()
#is null makes the data frame either true or false bassing on wheather the data is present or absent
```

```python
house_price_dataframe.isnull().sum()
#.sum() configures all the dataframes and tells the #.of features are false
```

```python
house_price_dataframe.describe()
```

```python
#finding the correlation the between the features
#so that we can do the dimensional reduction and make the model efficient
#let x and y be two features then
# positive correlation--> if x increases then y also increases
# negative correlation -->  if x increases the y decreases
correlation=house_price_dataframe.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# weights are generated between 0 and 1 by measuring the dependency
X=house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
print(X)
print(Y)
#making the features/inputs(x) and target function/output(y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
# here we have taken 80%-> training data 20%->testing size
print(X.shape,X_train.shape,X_test.shape)
```

```python
#training the model
model=XGBRegressor()
model.fit(X_train, Y_train)
```

```python
#predicting the values
training_data_prediction=model.predict(X_train)
print(training_data_prediction)
```

```python
#finding the r squared error -->loss function
score_1=metrics.r2_score(Y_train,training_data_prediction)
#finding the mean absolute error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
#errors shuld be less than 2 for better results
```

```python
#plot graph for that
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual vs Prediction")
plt.show()
```

```python
#predictions on the test data
test_data_prediction = model.predict(X_test)
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
